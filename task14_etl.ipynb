```python
import pandas as pd
import numpy as np
import sqlite3
import os
```

```python
os.makedirs("raw", exist_ok=True)
os.makedirs("processed", exist_ok=True)
os.makedirs("output", exist_ok=True)

print("Folders created!")
```

```python
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("superstore.csv", encoding='latin1')
df.head()
```

```python
df.to_csv("raw/raw_data.csv", index=False)

print("Raw dataset saved!")
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Remove missing values
df = df.dropna()

print("Data cleaned!")
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")

df.head()
```

```python
df["profit_margin"] = df["profit"] / df["sales"]

df["high_value_customer"] = df["sales"] > 1000

df.head()
```

```python
df.to_csv("processed/processed_data.csv", index=False)

print("Processed dataset saved!")
```

```python
customers = df[["customer_name", "segment"]].drop_duplicates()

orders = df[["order_id", "customer_name", "sales", "profit"]]

products = df[["product_name", "category", "sub-category"]].drop_duplicates()

print("Tables created!")
```

```python
conn = sqlite3.connect("database.sqlite")

customers.to_sql("customers", conn, if_exists="replace", index=False)
orders.to_sql("orders", conn, if_exists="replace", index=False)
products.to_sql("products", conn, if_exists="replace", index=False)

conn.close()

print("SQLite database created!")
```

```python
print("Original rows:", len(df))
print("Customers:", len(customers))
print("Orders:", len(orders))
print("Products:", len(products))
```
